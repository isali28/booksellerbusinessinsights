```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/archive.zip')
df.head()
```

```python
import sqlite3

# Create a connection to a new SQLite database
conn = sqlite3.connect('books.db')

# Load the DataFrame into a SQL table named 'bestsellers'
df.to_sql('bestsellers', conn, if_exists='replace', index=False)
```

```python
# Retrieve the top 5 bestselling books
query = """
 SELECT author, AVG(NULLIF(price, 0)) AS avg_price
 FROM bestsellers
 GROUP BY author
 HAVING AVG(NULLIF(price, 0)) IS NOT null
 ORDER BY avg_price DESC ;
"""

result = pd.read_sql_query(query, conn)
result


```

```python
result.to_csv("Author by AVG PRICE.csv", index=False)

# Download the CSV file
from google.colab import files
files.download("Author by AVG PRICE.csv")
```
